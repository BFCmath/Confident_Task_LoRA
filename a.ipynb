```python
a = "results_wrong_only.csv"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(a)
```

```python
# remove predicted_answer that are NaN
df = df[df["predicted_answer"].notna()]
df[["index","ground_truth","predicted_answer"]].to_csv("results_wrong_only_slimmed.csv", index=False)
```

```python
import pandas as pd
df2 = pd.read_csv("results_wrong_only_slimmed.csv")
```

```python
from grader import math_equal   
```

```python
df2["is_correct"] = df2.apply(
    lambda row: math_equal(row["predicted_answer"], row["ground_truth"]),
    axis=1
)
```

```python
df_correct = df2[df2["is_correct"] == False].copy()

df_correct.head()
```

```python
df_correct.to_csv("results_wrong_only_slimmed_checked.csv", index=False)
```
